```python
mean = df['Suhu Udara'].mean()
median = df['Suhu Udara'].median()
modus = df['Suhu Udara'].mode()[0]
print('mean =', mean)
print('median =', median)
print('modus=', modus)
```

```python
mean = df['Curah Hujan'].mean()
median = df['Curah Hujan'].median()
modus = df['Curah Hujan'].mode()[0]
print('mean =', mean)
print('median =', median)
print('modus=', modus)
```

```python
# Visualisasi dengan Boxplot
plt.figure(figsize=(10, 6))
plt.subplot(1, 2, 1)
plt.boxplot(df['Kelembapan Udara'])
plt.title('Boxplot Kelembapan Udara')

plt.subplot(1, 2, 2)
plt.boxplot(df['Curah Hujan'])
plt.title('Boxplot Curah Hujan')

plt.tight_layout()
plt.show()

# Visualisasi dengan Scatterplot
plt.figure(figsize=(6, 6))
plt.scatter(df['Kelembapan Udara'], df['Curah Hujan'])
plt.xlabel('kelembapan Udara')
plt.ylabel('Curah Hujan')
plt.title('scatterplot kelembapan udara vs Curah Hujan')
plt.show()
```

```python
# Plot histogram untuk curah hujan
plt.figure(figsize=(8, 6))
plt.hist(df['Curah Hujan'], bins=10, edgecolor='black')
plt.xlabel('Curah Hujan')
plt.ylabel('frekuensi')
plt.title('Histogram Curah Hujan')

# Tambahkan garis vertikal untuk menunjukkan nilai mean
mean_curah_hujan = df['Curah Hujan'].mean()
plt.axvline(mean_curah_hujan, color='black', linestyle='dashed', linewidth=1, label='Mean')
plt.text(mean_curah_hujan+1, 20, f'Mean: {mean_curah_hujan:.2f}', color='yellow')

# Tambahkan garis vertikal untuk menunjukkan nilai standar deviasi
std_curah_hujan = df['Curah Hujan'].std()
plt.axvline(mean_curah_hujan+std_curah_hujan, color='black',linestyle='dashed',linewidth=1, label='Std Dev')
plt.axvline(mean_curah_hujan-std_curah_hujan, color='black',linestyle='dashed',linewidth=1)
plt.text(mean_curah_hujan+std_curah_hujan+1, 15, f'Std Dev: {std_curah_hujan:.2f}', color='black')

plt.legend()
plt.show()

# Plot histogram untuk suhu
plt.figure(figsize=(8, 6))
plt.hist(df['Suhu Udara'], bins=10, edgecolor='black')
plt.xlabel('Suhu')
plt.ylabel('frekuensi')
plt.title('Histogram Suhu')

# Tambahkan garis vertikal untuk menunjukkan nilai mean
mean_suhu = df['Suhu Udara'].mean()
plt.axvline(mean_suhu, color='red', linestyle='dashed', linewidth=1, label='Mean')
plt.text(mean_suhu+1, 20, f'Mean: {mean_suhu:.2f}', color='red')

# Tambahkan garis vertikal untuk menunjukkan nilai standar deviasi
std_suhu = df['Suhu Udara'].std()
plt.axvline(mean_suhu+std_suhu, color='green', linestyle='dashed', linewidth=1, label='Std Dev')
plt.axvline(mean_suhu-std_suhu, color='green', linestyle='dashed', linewidth=1)
plt.text(mean_suhu+std_suhu+1, 15, f'Std Dev: {std_suhu:.2f}', color='green')

plt.legend()
plt.show()
```

```python
# Identifikasi pola dan anomali data

# Menampilkan informasi umum tentang dataset
print("Informasi Umum:")
print(df.info())
print()

# Melihat ringkasan statistik tentang dataset
print("Ringkasan Statistik:")
print(df.describe())
print()

# Menghitung korelasi antar variabel numerik
print("korelasi:")
correlation = df.corr()
print(correlation)
print()

# Menemukan anomali menggunakan metode IQR (Interquartile Range)
Q1 = df.quantile(0.25)
Q3 = df.quantile(0.75)
IQR = Q3 - Q1
outliers = ((df < (Q1 - 1.5 * IQR)) | (df > (Q3 + 1.5 * IQR))).sum()
print("Jumlah Anomali:")
print(outliers)
print()

# Menemukan nilai-nilai yang sering muncul (modus)
print("Modus:")
modus = df.mode().iloc[0]
print(modus)
print()
```
